```python
import math

class Conversion3D:
  def _init_(self, a,b,c,d=None):
    self.a=a
    self.b=b
    self.c=c
    self.d=d

  def coordRect(self):
    print("Ingrese el valor de X")
    x=int(input())
    print("Ingrese el valor de Y")
    y=int(input())
    print("Ingrese el valor de Z")
    z=int(input())
    return x, y, z

  def execute_with_error_handling(func_to_execute):
    try:
        # Ejecuta la función proporcionada como argumento
        result = func_to_execute()
        return result
    except ValueError as e:
        # Captura cualquier excepción que se produzca durante la ejecución de la función
        print(f"Se produjo un error: {e}")
        return None

  def coordCil(self):
    print("Ingrese el valor de rc")
    rc=int(input())
    print("Ingrese el valor de \u03C6")
    p=int(input())
    print("Ingrese el valor de Z")
    z=int(input())
    return rc, p, z


  def coordEsf(self):
    print("Ingrese el valor de re")
    re=int(input())
    print("Ingrese el valor de \u03C6")
    p=int(input())
    print("Ingrese el valor de \u03B8")
    t=int(input())
    return re,p,t

  def ntuplas(self):
    print("Ingrese el numero de tuplas que va a ingresar")
    nt=input()
    return(nt)

  def printNum(self,i,nt):
    print(f"Punto {i+1} de {nt}:")

  def rectToCil(self,i,nt):
    x, y, z = self.coordRect()
    rc=math.sqrt((x**2)+(y**2))
    p=math.atan2(y,x)
    p=(p*180)/math.pi
    z=float(z)
    print(f"Punto {i+1} de {nt}:")
    self.printNum(i,nt)
    print(f"(x, y, z) = ({x},{y},{z}) -----> (rc, \u03C6, z) = ({round(rc,1)}, {round(p,1)}, {round(z,1)})")

  def rectToEsf(self,i, nt):
    a,b,c=self.coordRect()
    re=math.sqrt((a**2)+(b**2)+(c**2))
    p=math.atan2(b,a)
    p=(p*180)/math.pi
    t=math.acos((c/re))
    t=(t*180)/math.pi
    self.printNum(i,nt)
    print(f"(x, y, z) = ({a},{b},{c}) -----> (rc, \u03C6, \u03B8) = ({round(re,1)}, {round(p,1)}, {round(t,1)})")

  def cilToRect(self,i, nt):
    a,b,c=self.coordCil()
    b=math.radians(b)
    x=a*math.cos(b)
    y=a*math.sin(b)
    z=c
    self.printNum(i,nt)
    print(f"(rc, \u03C6, z) = ({a},{b},{c}) -----> (x, y, z) = ({round(x,1)}, {round(y,1)}, {round(z,1)})")

  def cilToEsf(self,i, nt):
    a,b,c=self.coordCil()
    re=math.sqrt((a**2)+(c**2))
    p=b
    t=math.atan2(a,c)
    t=(t*math.pi)/180
    self.printNum(i,nt)
    print(f"(rc, \u03C6, z) = ({a},{b},{c}) -----> (re, \u03C6, \u03B8) = ({round(re,1)}, {round(p,1)}, {round(t,1)})")

  def esfToRect(self,i, nt):
    a,b,c=self.coordEsf()
    x=a*(math.sin(c))*(math.cos(b))
    y=a*(math.sin(c))*(math.sin(b))
    z=a*math.cos(c)
    self.printNum(i,nt)
    print(f"(re, \u03C6, \u03B8) = ({a},{b},{c}) -----> (x, y, z) = ({round(x,1)}, {round(y,1)}, {round(z,1)})")

  def esfToCil(self,i, nt):
    a,b,c=self.coordEsf()
    rc=a*math.sin(c)
    p=b
    z=a*math.cos(b)
    self.printNum(i,nt)
    print(f"(re, \u03C6, \u03B8) = ({a},{b},{c}) -----> (rc, \u03C6, z) = ({round(rc,1)}, {round(p,1)}, {round(z,1)})")

print("Sus coordenadas están en formato: \n1. Rectangular\n2. Cilindrico\n3. Esférico")

c3D=Conversion3D()
d=input()

if d=="1":
  print("Desea convertir a:\n1. Cilindrico\n2. Esferico")
  conv=input()

  if conv=="1":
    nt=int(c3D.ntuplas())

    for i in range(nt):
      c3D.execute_with_error_handling(c3D.rectToCil(i,nt))
      #c3D.rectToCil(i,nt)

  if conv=="2":
    nt=int(c3D.ntuplas())

    for i in range(nt):
      c3D.rectToEsf(i, nt)

if d=="2":
    print("Desea convertir a:\n1. Rectangular\n2. Esferico")
    conv=input()

    if conv=="1":
      nt=int(c3D.ntuplas())

      for i in range(nt):
        c3D.cilToRect(i, nt)

    if conv=="2":
      nt=int(c3D.ntuplas())

      for i in range(nt):

        c3D.cilToEsf(i, nt)

    if d=="3":
      print("Desea convertir a:\n1. Rectangular\n2. Cilindrico")
      conv=input()

      if conv=="1":
        nt=int(c3D.ntuplas())

        for i in range(nt-1):
          c3D.esfToRect(i, nt)

      if conv=="2":
        nt=int(c3D.ntuplas())

        for i in range(nt-1):
          c3D.cilToEsf(i, nt)
```
